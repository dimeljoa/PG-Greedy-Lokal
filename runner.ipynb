```python
import os
import sys
import subprocess
from pathlib import Path

# Pfade
in_path = Path("C:/Users/dimel/Downloads/dataset (1)/dataset")
exe = Path("./build/csv_labeler.exe")
out_path = Path("./build/results")

# Prüfe, ob Executable existiert
if not exe.exists():
    print("Executable not found. Please build the project first.")
    sys.exit(1)

# Falls der Ausgabeordner nicht existiert, erstelle ihn
out_path.mkdir(parents=True, exist_ok=True)

# Liste alle CSV-Dateien im Eingabeordner auf
csv_files = list(in_path.glob("*.csv"))
if not csv_files:
    print(f"Keine CSV-Dateien gefunden in: {in_path}")
    sys.exit(0)

# Verarbeite jede CSV-Datei
for csv_file in csv_files:
    out_file = out_path / csv_file.name
    print(f"\nProcessing: {csv_file.name}")
    print(f"   Output -> {out_file}")

    # Befehl mit vollständigen Pfaden
    cmd = [str(exe), str(csv_file), str(out_file)]

    try:
        # subprocess.run zeigt stdout und stderr live
        subprocess.run(cmd, cwd=exe.parent, check=True)
        print("Done.")

    except subprocess.CalledProcessError as e:
        print(f"Fehler beim Ausführen von {csv_file.name}: {e}")

print("\nAlle Dateien verarbeitet.")
```

```python
import os
import sys
import subprocess
from pathlib import Path

in_path = Path("C:/Users/dimel/Downloads/dataset (1)/dataset")
exe = Path("./build/csv_labeler.exe")
out_path = Path("./build/results")

if not exe.exists():
    print("❌ Executable not found. Please build the project first.")
    sys.exit(1)

out_path.mkdir(parents=True, exist_ok=True)
csv_files = list(in_path.glob("*.csv"))

if not csv_files:
    print(f"⚠️ Keine CSV-Dateien gefunden in: {in_path}")
    sys.exit(0)

for csv_file in csv_files:
    out_file = out_path / csv_file.name
    print(f"\n📄 Processing: {csv_file.name}")
    print(f"   Output -> {out_file}")

    cmd = [str(exe), str(csv_file), str(out_file)]

    # Starte Prozess mit direkter Ausgabe
    process = subprocess.Popen(
        cmd,
        cwd=exe.parent,
        stdout=sys.stdout,
        stderr=sys.stderr,
        text=True,
        bufsize=1  # Zeilenweise Ausgabe
    )

    process.wait()
    if process.returncode == 0:
        print("✅ Done.")
    else:
        print(f"❌ Fehler bei {csv_file.name}")
```
